# Car price predication: Data Visualization

```python
car.columns= ["Year	","Selling_Price","Present_Price",
               "Driven_kms","Car_Name","Fuel_Type","Selling_type","Transmission","Owner"]
plt.title("Car price predication")
sns.histplot(x="Car_Name", hue="Fuel_Type", data=car)
plt.show()
```

```python
sns.countplot(x = 'Fuel_Type',data=car)
plt.show()
```

```python
sns.scatterplot(x='Selling_Price', y='Present_Price',
                hue='Fuel_Type',
                data=car)
plt.show()
```

```python
#Pairs plot\matrix plot

sns.pairplot(car)
plt.show()
```

```python
sns.scatterplot(x='Selling_Price', y='Present_Price',
                hue='Transmission',
                data=car)
plt.show()
```

```python
car.plot(kind="box")
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(3, 3))
sns.heatmap(car.corr())
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize = (5,5)) 
corr = car.corr().iloc[:,-1:] 
sns.heatmap(corr, annot = True) 
plt.show() 
```
